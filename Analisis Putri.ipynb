```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

housing = fetch_california_housing()
dataset = pd.DataFrame(data=housing.data, columns=housing.feature_names)
dataset['target'] = housing.target
print(dataset.head())
```

```python
X = dataset.drop('target',axis=1)
Y = dataset["target"]
```

```python
print(X.corr())
corr_value = round(dataset['MedInc'].corr(dataset['HouseAge']),2)
```

```python
import statsmodels.api as sm
statsModel = sm.OLS(Y,X).fit()
print(statsModel.summary())
```

```python
###Start code here
r_squared = round(statsModel.rsquared,2)
###End code(approx 1 line)
```

```python
import hashlib
import pickle
import os

def gethex(ovalue):
    hexresult=hashlib.md5(str(ovalue).encode())
    return hexresult.hexdigest()
def pickle_ans1(value):
    hexresult=gethex(value)
    with open('ans/output1.pkl', 'wb') as file:
        print(hexresult)
        pickle.dump(hexresult,file)
def pickle_ans2(value):
    hexresult=gethex(value)
    with open('ans/output2.pkl', 'wb') as file:
        print(hexresult)
        pickle.dump(hexresult,file)

# Create the 'ans' directory if it doesn't exist
os.makedirs('ans', exist_ok=True)

pickle_ans1(corr_value)
pickle_ans2(r_squared)
```
